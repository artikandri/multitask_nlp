Creating `text_data.csv` and `annotations.csv` for Measuring Hate Speech dataset.

```python
%load_ext autoreload
%autoreload 2

import os
import sys
sys.path.append('../../../')

import pandas as pd
from tqdm.auto import tqdm

from datasets import load_dataset
from multitask_nlp.settings import DATASETS_DIR

tqdm.pandas()

dataset_path = DATASETS_DIR / 'measuring_hate_speech'

isExist = os.path.exists(dataset_path)

if not isExist:
    os.makedirs(dataset_path)
    print("The new directory is created!")
```

```python
dataset = load_dataset("ucberkeley-dlab/measuring-hate-speech")
```

```python
columns_to_preserve = ['comment_id', 'annotator_id', 'platform', 'sentiment', 'respect',
       'insult', 'humiliate', 'status', 'dehumanize', 'violence',
       'genocide', 'attack_defend', 'hatespeech', 'hate_speech_score',
       'text']

df = df[columns_to_preserve]
```

```python
texts_df = df[['comment_id', 'text']]
texts_df = texts_df.rename(columns={'comment_id': 'text_id'})
texts_df = texts_df.drop_duplicates()
```

```python
texts_df
```

```python
annotations_df = df.copy()
annotations_df = annotations_df.drop(['platform', 'text'], axis=1)
annotations_df = annotations_df.rename(columns={'comment_id': 'text_id'})
```

```python
texts_df.to_csv(dataset_path / 'data.csv', index=False)
annotations_df.to_csv(dataset_path / 'annotations.csv', index=False)
```
